```python
import re
import pandas as pd, numpy as np
```

```python
# Downloaded from https://instances.vantage.sh - 2021-06-15
df = pd.read_csv("Amazon EC2 Instance Comparison.csv")

def parse_float(pattern):
    p = re.compile(pattern)
    def fn(s):
        r = p.match(s)
        return float(r.group(1)) if r is not None else np.nan
    return fn
    
df['linux_od_cost_usd_hr'] = df['Linux On Demand cost'].apply(parse_float(r"[$]([0-9]+\.[0-9]+) hourly"))
df['vcpus'] = df['vCPUs'].apply(parse_float(r"([0-9]+) vCPUs"))
df['vfreq_ghz'] = df['Clock Speed(GHz)'].apply(parse_float(r"([0-9]+\.[0-9]+) GHz"))
df['mem_gb'] = df['Memory'].apply(parse_float(r"([0-9]+\.[0-9]+) GiB"))

df['mem_per_vcpu'] = df['mem_gb'] / df['vcpus']
df['cost_per_vcpu_hr'] = df['linux_od_cost_usd_hr'] / df['vcpus']

df = df[['API Name', 'Physical Processor', 'linux_od_cost_usd_hr', 'vcpus', 
         'vfreq_ghz', 'mem_gb', 'mem_per_vcpu', 'cost_per_vcpu_hr']]

df.head()
```

```python
df = df[~df.cost_per_vcpu_hr.isnull()]
df = df[df.vfreq_ghz > 3.0]
```
